```python
def get_feature_vec(input_json):
    all_list = []
    for idxx, each in enumerate(input_json):
        
        try:
            raw = tokenize_and_normalize(each['input'])
            r_len = len(raw)*1.0
        except:
            import pdb; pdb.set_trace()
        each_data_fea  = []

        if r_len ==0:
            continue
        if len(raw)<cutoff_start or len(raw)>cutoff_end:
            continue

        # each_data_fea  = [len(raw) / 100.]
        
        each_data_fea = [ind_d / r_len for ind_d in each['avg_common_features']]
        for ek in each['common_features'].keys():
            each_data_fea.extend([ind_d / r_len for ind_d in each['common_features'][ek]])
        
        each_data_fea.extend([ind_d / r_len for ind_d in each['common_features_ori_vs_allcombined']])

        for ek in each['fzwz_features'].keys():
            each_data_fea.extend(each['fzwz_features'][ek])

        all_list.append(np.array(each_data_fea))

        if idxx == 400:
            break


    all_list = np.vstack(all_list)

    return all_list
```

### Calculate editing distance

```python
rewrite_h_equi_json_list = load_json_list('rewrite_h_equi_json_list.json')
rewrite_ai_equi_json_list = load_json_list('rewrite_ai_equi_json_list.json')
```

```python
def equi_data_generate(json_list: list) -> list:
    for data in json_list:
        original = data['input']

        bow_dct = {}
        all_bow_dct = [0 for i in range(NGRAM_NUM)]
        cnt = 0        

        lev_dct = {}
        whole_combined=''
        
        data_tkeys = [key for key in data.keys() if key.__contains__('tmp&')]
        data_rkeys = [key for key in data.keys() if key.__contains__('final*')]

        for transformed, reversal in zip(data_tkeys, data_rkeys):
            # whole_combined += (' ' + data[prompt_inst])

            # compute bag of words across the n-grams
            bow_cls = select_edit_method('bow')(original, data[reversal])
            b_res = bow_cls.calculate_sentence_common()
            bow_dct[reversal] = b_res 
            all_bow_dct = bow_cls.sum_for_list(all_bow_dct, b_res)

            # compute levenshtein distance 
            lev_cls = select_edit_method('lev')(original, data[reversal])
            l_res = lev_cls.calculate_lev_distance()
            lev_dct[reversal] = l_res 

            cnt += 1
        
        data['fzwz_features'] = lev_dct
        data['common_features'] = bow_dct
        data['avg_common_features'] = [a/cnt for a in all_bow_dct]
        data['common_features_ori_vs_allcombined'] = select_edit_method('bow')(original, whole_combined).calculate_sentence_common()

    return json_list
```
